Usando los métodos estudiados, determine todas las raíces de la ecuación:

$$x5−34.5∗x4+384.5∗x3−1559∗x2+2646∗x−1612=0.$$

```python
from scipy.optimize import bisect
import matplotlib.pyplot as plt
import numpy as np

def funcion(x: float):
    return x**5 - 34.5*x**4 + 384.5*x**3 - 1559*x**2 - 2646*x + 1612

x = np.linspace(-10, 10, 100)

plt.plot(x , [funcion(i) for i in x])
plt.xlim(-2.5, 1)
plt.ylim(-0.1, 0.1)
plt.axhline(0, color='gray', linewidth=0.8)
plt.axvline(0, color='gray', linewidth=0.8)
plt.show()
```

```python
# Raiz 1
root = bisect(f = funcion, a = -2, b = 0)
print(root)

# Raiz 2
root2 = bisect(f = funcion, a = 0, b = 2)
print(root2)

#  Raiz 3
root3 = bisect(f = funcion, a = -1, b = -2.5)
print(root3)
```
